반복문은 일반적으로 for, in 키워드를 사용하여 작성한다.

for 변수 in 객체  의 형태로 작성되며 객체는 일반적으로 iterable 한 데이터 타입을 활용한다.

```python
gems = ["ruby", "sapphire", "emerald", "carnelian"]

for gem in gems:
    print(gem, len(gem))
```

```python
for word in "rubydango":
    print(word)
```

```python
for x in {
    "ruby" : "루비",
    "sapphire" : "사파이어",
    "emerald" : "에메랄드",
    "carnelian" : "카넬리안"
}:
    print(x)
    
#딕셔너리는 일반적으로 키를 뽑아온다.

for x, y in {
    "ruby" : "루비",
    "sapphire" : "사파이어",
    "emerald" : "에메랄드",
    "carnelian" : "카넬리안"
}.items():
    print(x, y)
```

```python
for [x, y] in [[1, 2], [3, 4], [5, 6]]:	
    print(x, y) 

#리스트 안의 리스트(중첩 리스트)도 반복문을 이용해 출력 가능하다.
```

```python
a = 'abcde'
b = [val + 'k' for val in a]

print(b)
```

for 변수 in range(반복횟수):
    (실행 구문)

로 몇 번 반복문을 실행할 것인지 결정할 수 있다.

```python
for i in range(10):
    print(i)
```

```python
for i in range(5, 9):
    print(i)

# range에 따른 반복의 시작지점과 끝나는 지점을 지정할 수 있다.

for i in range(5, 15, 3):
    print(i)

# 몇 번마다 반복할 것인지도 지정 가능.
```

```python
introduce = ['i', 'am', 'a', 'ruby']

for idx in range(len(introduce)):
    print(idx, introduce[idx])
```

```python
sum(range(11))
```

```python
list = []

for n in range(1,11):
    list.append(n * n)

print(list)

#print([n ** 2 for n in range(1, 11)])

#위의 주석은 리스트 내포를 활용해 작성하였다.
```

while로도 실행구문을 반복시킬 수 있다.

```python
i = 1

while i <= 10:
    print(i)
    i = i + 1
```
