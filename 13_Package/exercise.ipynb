패키지란 모듈이 모여있는 집합이다. 다시 말해 폴더와 같다.

```python
import gems.gem.ruby_module

print(gems.gem.ruby_module.gems)
```

모듈에서와 같이 as 키워드를 사용해 작성을 단축시킬 수 있다.

```python
import gems.gem.ruby_module as ruby

print(ruby.gems)
```

```python
from gems.gem import ruby_module

print(ruby_module.gems)
```
