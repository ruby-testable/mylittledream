ğŸ§© ë¬¸ì œ 1 â€“ í•™ìƒ ì ìˆ˜ ê´€ë¦¬ í´ë˜ìŠ¤

Student í´ë˜ìŠ¤ë¥¼ ë§Œë“¤ê³  ì†ì„±: name, scores(ë¦¬ìŠ¤íŠ¸)

ë©”ì„œë“œ:

í•™ìƒ ê°ì²´ë¥¼ ìƒì„±í•˜ê³  ì ìˆ˜ë¥¼ ì¶”ê°€í•œ ë’¤, í‰ê· ê³¼ í•™ì ì„ ì¶œë ¥í•˜ì„¸ìš”.

```python
class Student:

    def __init__(self, name, scores):
        self.name = name
        self.scores = scores

    def average(self):
        return sum(self.scores) / len(self.scores)

    def grade(self):
        avg = self.average()
        if avg >= 90:
            return "A"

        elif avg >= 80:
            return "B"
        
        elif avg >= 70:
            return "C"
        
        elif avg >= 60:
            return "D"
        
        else:
            return "E"

    def add_score(self, score):
        return self.scores.append(score)

my_grade = [80, 90, 95, 100]
doyoung = Student("ê¹€ë„ì˜", my_grade)
doyoung.add_score(50)

print(doyoung.average())
print(doyoung.grade())
```

ğŸ”¹ ë¬¸ì œ 4 â€“ ëª¨ìŒ/ììŒ ì¹´ìš´íŠ¸ í´ë˜ìŠ¤

TextAnalyzer í´ë˜ìŠ¤

ì†ì„±: text

ë©”ì„œë“œ:

1. count_vowels() â†’ ëª¨ìŒ(a,e,i,o,u) ê°œìˆ˜ ë”•ì…”ë„ˆë¦¬ ë°˜í™˜

2. count_consonants() â†’ ììŒ ê°œìˆ˜ ë”•ì…”ë„ˆë¦¬ ë°˜í™˜

ëŒ€ì†Œë¬¸ì êµ¬ë¶„ ì—†ì´ ì²˜ë¦¬, ê³µë°± ë° ìˆ«ì ì œì™¸

```python
class TextAnalyzer:

    def __init__(self, text):
        self.text = text.lower()

    def count_vowels(self):
        vowels = "aeiou"
        return {vowel: self.text.count(vowel) for vowel in vowels}
    
    def count_consonants(self):
        consonants = "bcdfghjklmnpqrstvwxyz"
        return {consonant: self.text.count(consonant) for consonant in consonants}

ruby = TextAnalyzer("sapphire is best gem in the world")

print(ruby.count_vowels())
print(ruby.count_consonants())
```
