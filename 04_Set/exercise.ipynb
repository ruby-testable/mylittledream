add는 하나의 값만 추가 가능한 반면, update 메소드를 활용하면 한번에 여러 개의 값을 추가하는 것이 가능하다.

셋은 집합과 유사한 면이 있으므로 수학의 집합 연산을 실행 할 수 있다.

union 메서드를 활용해 합집합을 형성할 수 있다.

```python
javaTeam = {'Squirrel','Gorilla', 'Tiger', 'Sheep', 'monkey', 'wolf'}
pythonTeam = {'Pig', 'Squirrel', 'Bear', 'Gorilla'}

ohgiraffers = javaTeam.union(pythonTeam)
# union 메서드를 전후로 두 개의 집합을 합치는 것

print(javaTeam | pythonTeam)   
print(ohgiraffers)  

"""
| << 바 형태의 명령어는 or와 같은 의미를 가진다.
"""
```

intersection 메서드나 & 연산자를 활용해 교집합을 형성할 수 있다.

```python
print(javaTeam & pythonTeam)                # {'Squirrel', 'Gorilla'}
print(javaTeam.intersection(pythonTeam))    # {'Squirrel', 'Gorilla'}
```

difference 메서드나 - 연산자를 활용해 차집합을 형성할 수 있다.

```python
print(javaTeam - pythonTeam)              # {'monkey', 'wolf', 'Sheep', 'Tiger'}
print(javaTeam.difference(pythonTeam))    # {'monkey', 'wolf', 'Sheep', 'Tiger'}

#차집합의 형성 시에는 왼쪽에서 오른쪽을 빼는 형태로 만들어진다. 순서가 중요하다고 볼 수 있다.
```

symmetric_difference 메서드나 ^ 메서드를 활용해 대칭 차집합(합집합 - 교집합)을 형성할 수 있다.

```python
print(javaTeam ^ pythonTeam)                        
print(javaTeam.symmetric_difference(pythonTeam))    
```

copy 메서드를 활용해 해당 셋을 복사하여 생성할 수 있다.

```python
ohgiraffers = {'Squirrel','Gorilla', 'Tiger', 'Sheep', 'monkey', 'wolf'}
zzap_ohgiraffers = ohgiraffers.copy()
print(ohgiraffers)         # {'wolf', 'Sheep', 'Tiger', 'Gorilla', 'monkey', 'Squirrel'}
print(zzap_ohgiraffers)    # {'wolf', 'Sheep', 'Tiger', 'Gorilla', 'monkey', 'Squirrel'}

zzap_ohgiraffers.remove('Tiger')
print(ohgiraffers)         # {'wolf', 'Sheep', 'Tiger', 'Gorilla', 'monkey', 'Squirrel'}
print(zzap_ohgiraffers)    # {'wolf', 'Sheep', 'Gorilla', 'monkey', 'Squirrel'}
```

딕셔너리는 여러 개의 자료를 하나로 묶은 자료라는 점에서 리스트 등과 공통점이 있다.

딕셔너리는 키와 값의 쌍으로 구성된 자료형으로, 키를 통해 값을 찾을 수 있는 것에 의의가 있다.

예시) 딕셔너리명 = { 키1: 값1, 키2: 값2, ... }

```python
teacher = {'name': 'bear', 'team': 'ohgiraffers'}

print(type(teacher))    # <class 'dict'>

# key 값을 활용한 value 탐색
print(teacher['name'])    # bear
print(teacher['team'])    # ohgiraffers

#리스트나 튜플에서는 인덱스(숫자)를 입력해야 했던 반면, 딕셔너리는 키를 입력하는 것으로 조회가능
```

```python
ohgiraffers = {
    'tiger' : 'langChain', 
    'Squirrel' :'JPA', 
    'Bear':'Python', 
    'Gorilla':'Spring'
}
#위의 표기 방식은 가독성을 더 추구한 방식이라 볼 수 있을 것이다.

print(ohgiraffers)   
print(ohgiraffers['tiger'])      
print(ohgiraffers['Squirrel'])   
```

키는 해당 딕셔너리 내에서 고유하고 불변해야 하는 데이터 타입이다. (값은 중복 가능)

이 때문에 키로 가변 자료형인 리스트나 딕셔너리는 사용이 불가능한 반면,

튜플은 사용 가능하다.

```python
ohgiraffers = {
    'tiger':'langChain', 
    'Squirrel':'JPA', 
    'Bear':'Python', 
    'Gorilla':'Spring', 
    "Squirrel":"MyBatis"
}
print(ohgiraffers['Squirrel'])
# 키 Squirrel에 해당하는 값이 MyBatis로 바뀐 것을 확인할 수 있다.

# invalid_key_teacher = {["name", "team"] : 'bear.ohgiraffer'}  # Error 발생

valid_key_dic = {("name", "team") : 'bear.ohgriaffers'}
print(valid_key_dic[("name", "team")])    # bear.ohgriaffers
```

zip 메서드를 활용해 두 가지의 시퀀스형 자료를 딕셔너리로 묶을 수 있다.

```python
k = ['one', 'two', 'three']
v = [1,2,3]

# 두개의 자료를 순서대로 쌍으로 묶은 zip객체를 반환
t = zip(k,v)
print(t) #<zip object at 0x034ABE90>
print(type(t)) #<class 'zip'>

for k,v in t:
    print(k,v)
```

딕셔너리를 복사할 수 있다.

이 때, 얕은 복사와 깊은 복사의 개념으로 나누어진다.

```python
# 1.shallow copy
a = {'a': 1, 'b': 2, 'c': 3}
b = a # 사전 레퍼런스의 복사 (공유됨)
print(a)
print(b)

a['a'] = 100
print(a)
print(b)

"""
딕셔너리 b의 키 a에 해당하는 값이 동일한 이유는 딕셔너리 a, b가 같은 주소값을 참조하기 때문이다.
"""
```

```python
# 2.deep copy
c = {'a': 1, 'b': 2, 'c': 3}
d = c.copy()
print(c)
print(d)

c['a'] = 300
print(c)
print(d) # 변경되지 않음.
```

sorted 명령어를 활용해 딕셔너리를 정렬할 수 있다.

```python
# dict의 정렬 : sorted
numbers = {'first': 1, 'second': 2, 'third': 3, 'Fourth': 4}
numbers['hundred'] = 100
numbers['fifth'] = 5

print(list(numbers)) #['first', 'second', 'third', 'Fourth', 'hundred', 'fifth']

# key값 정렬
print(sorted(numbers)) #['Fourth', 'fifth', 'first', 'hundred', 'second', 'third']

# value값 정렬
print(sorted(numbers.values())) #[1, 2, 3, 4, 5, 100]

# key정렬 (value값 정렬에 따라)
print(sorted(numbers, key=numbers.__getitem__))

```
