### **문제 2: `.csv` 파일 생성 및 읽기**

다람쥐는 이직 준비를 위해 지원했던 회사들의 정보를 저장하려고 합니다.

1. `applications.csv` 파일을 생성하고 다음 데이터를 작성하세요.
    
    ```
    회사명,직무,결과
    회사A,데이터 분석가,합격
    회사B,소프트웨어 엔지니어,불합격
    회사C,제품 관리자,진행 중
    ```
    
2. 파일을 닫고 다시 열어 모든 데이터를 출력하세요.

````python
import csv

f = open("application.csv", "w", encoding = "utf-8")
f.write("""
회사명,직무,결과\n
회사A,데이터 분석가,합격\n
회사B,소프트웨어 엔지니어,불합격\n
회사C,제품 관리자,진행 중
""")
f.close()

with open("application.csv", "r", encoding = "utf-8") as csv_file:
    reader = csv.reader(csv_file)
    for row in reader:
        print(",".join(row))
````

````python
# csv 형태로 저장하기

import csv

data = [
    ["회사명", "직무", "결과"],
    ["회사A","데이터 분석가","합격"],
    ["회사B","소프트웨어 엔지니어","불합격"],
    ["회사C","제품 관리자","진행 중"]
]

with open("application.csv", "w", newline = "", encoding = "utf-8") as csv_file:
    writer = csv.writer(csv_file) # csv 작성 도구 생성
    writer.writerows(data) # 여러 행을 한꺼번에 작성

with open("application.csv", "r", encoding = "utf-8") as csv_file:
    reader = csv.reader(csv_file) # 파일 객체에 한 줄 씩 리스트로 넣는다는 뜻
    for row in reader:
        print(",".join(row)) # 리스트를 문자열로 변환하여 출력한다는 뜻
````

````python
# 딕셔너리 형태로 저장하기

import csv

data = [
    {"회사명": "회사A", "직무": "데이터 분석가", "결과": "합격"},
    {"회사명": "회사B", "직무": "소프트웨어 엔지니어", "결과": "불합격"},
    {"회사명": "회사C", "직무": "제품 관리자", "결과": "진행 중"}
]

with open("application.csv", "w", newline = "", encoding = "utf-8") as csv_file:
    fieldnames = ["회사명", "직무", "결과"]
    writer = csv.DictWriter(csv_file,fieldnames = fieldnames) 
    writer.writeheader()
    writer.writerows(data)

with open("application.csv", "r", encoding = "utf-8") as csv_file:
    reader = csv.DictReader(csv_file) # 파일 객체에 한 줄 씩 리스트로 넣는다는 뜻
    for row in reader:
        print(row) # 리스트를 문자열로 변환하여 출력한다는 뜻
````

### **문제 3: `.log` 파일 읽기**

다람쥐는 프로그램 실행 로그를 확인하려고 합니다.

1. `system.log` 파일을 다음 내용으로 생성하세요.
    
    ```
    [INFO] 2024-01-01 12:00:00: 시스템 시작
    [WARN] 2024-01-01 12:05:00: 메모리 사용량 증가
    [ERROR] 2024-01-01 12:10:00: 프로그램 종료
    ```
    
2. 파일을 읽어와 `WARN` 또는 `ERROR`가 포함된 줄만 출력하세요.

````python
f = open("system.log", "w", encoding = "utf - 8")
f.write("[INFO] 2024-01-01 12:00:00: 시스템 시작\n[WARN] 2024-01-01 12:05:00: 메모리 사용량 증가\n[ERROR] 2024-01-01 12:10:00: 프로그램 종료")
f.close

with open('system.log', 'r', encoding = "utf - 8") as f:
    for line in f:
        if "WARN" in line or "ERROR" in line:
            print(line)
````

다람쥐는 좋아하는 음식 데이터를 저장하고 싶습니다.

1. `favorite_foods.json` 파일을 생성하고 다음 데이터를 작성하세요.
    
    ```json
    {
        "name": "다람쥐",
        "favorite_foods": ["도토리", "호두", "밤"]
    }
    ```
    
2. 파일을 읽어와 다람쥐의 좋아하는 음식 리스트를 출력하세요.

````python
with open("favorite_foods.json", "w", encoding = "utf-8") as json_file:
    json_file.write('{"name":"루비", "favorite_foods":["씨없는수박", "샤인머스크"]}')

with open("favorite_foods.json", "r", encoding = "utf-8") as json_file:
    json_text = json_file.read()
    print(json_text.split("[")[1].split("]")[0])
````
