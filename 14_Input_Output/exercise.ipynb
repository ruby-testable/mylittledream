```python
f = open("dev/ruby_gem.txt", "r", encoding = "utf-8")

while True:
  line = f.readline()
  
  if not line:
    break
  print(line)
  
f.close()
```

```python
f = open("dev/ruby_gem.txt", "r", encoding = "utf-8")

lines = f.readlines()
for line in lines:
    print(line)

f.close()
```

```python
f = open("dev/ruby_gem.txt", "r", encoding = "utf-8")

for line in f:
    print(line)

f.close()
```

파일에 내용을 덮어쓰지 않고 추가만 해줄수도 있다. 

이 경우에는 "w"가 아닌 "a"로 파일을 열어야 한다.

```python
f1 = open("dev/ruby_gem.txt", "a", encoding = "utf-8")
f1.write("!")
f1.close()

f2 = open("dev/ruby_gem.txt", "r", encoding = "utf-8")
data = f2.read()
print(data)
f2.close()
```
