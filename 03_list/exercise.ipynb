del 키워드를 통해 원본 배열의 일부 요소 또는 전체 목록을 제거할 수 있다.

```python
abclist = ['A','B','C','D','E','F']
print(abclist)    # ['A', 'B', 'C', 'D', 'E', 'F']

del abclist[0]    # 0번째 인덱스의 요소 제거
print(abclist)    # ['B', 'C', 'D', 'E', 'F']

del abclist[1:3]  # [시작 : 끝-1]
print(abclist)    # ['B', 'E', 'F']

del abclist[:]    # 모든 요소 제거
print(abclist)    # []
```

여러 개의 값을 하나의 데이터 구조로 묶어서 관리할 수 있는 불변(immutable) 시퀀스이다. 리스트와 유사하지만, 한번 생성되면 각 요소 값을 생성하거나 수정, 삭제할 수 없다. 주로 데이터의 집합을 안전하게 유지하거나, 함수에서 여러 값을 반환할 때 사용된다.

```python
# 튜플의 요소는 리스트와는 달리 대괄호가 아니라 소괄호를 사용해 표현한다.

tupl = (1, 2, 'hello'), ('test', 1,2,3,4)
print(tupl)          
print(tupl[0])      
print(tupl[0][0])    
print(tupl[1][0])
# 집합 내부의 집합을 원소로 취급할지, 집합으로 취급할지 선택할 수 있는 것처럼 보인다.
```

소괄호로 감싸지 않아도, 하나의 변수에 여러 개의 값을 할당함으로써 튜플을 생성할 수 있다.

```python
x1 = 1, 2, 3
print(x1)
print(id(x1))

x = 2, 3, 4
print(x)
print(id(x))

print(type(x1))
```

in 함수를 이용해 시퀀스 자료 내부에 찾고자 하는 자료가 있는지 확인 할 수 있다.

```python
print('대한민국' in address)       # True
print('국민한대' in address)       # False
print('서울시서초구' in address)    # False, 띄어쓰기가 없다.

print('광주광역시' in location)                   # True
print(['광주광역시', '부산광역시'] in location)    # False
print(['서울특별시', '부산광역시'] in location)    # False

"""
['광주광역시', '부산광역시']에 대한 in 함수 사용시 false가 출력되는 이유는 앞의 것이
하나의 원소로 취급되기 때문이다.
"""

nation = ('대한민국', '미국', '중국', '일본', '프랑스', ('대한민국', '미국'))
print('미국' in nation)                 # True
print(('대한민국', '미국') in nation)
print(nation[5][1])
```
