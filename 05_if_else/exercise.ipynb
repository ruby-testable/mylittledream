조건문을 활용해 조건이 맞는지 검사하고 True or False 결과에 따라 다른 결과를 도출 할 수 있다.

```python
my_name = "김도영"

if my_name == "김도영" :
    print(my_name + "은 인간이다.")
```

if 는 True 일때 실행, else 는 False 일 때 실행되는 구문을 지정한다.

```python
favorite_gem = "ruby"

if favorite_gem == "ruby":
    print("루비")

else:
    print("다른 거")
```

elif 구문을 활용해 몇 가지 조건을 단계적으로 검증해 결과를 출력할 수 있다.

```python
favorite_gem = "ruby"

if favorite_gem == "ruby" :
    print("루비 최고")

elif favorite_gem[0] == "c":
    print("carnelian")

else:
    print("몰라레후")

    #elif 는 한 개 이상 사용 가능하다.
```

```python
print("참" if favorite_gem == "ruby" else "거짓")
print("참" if favorite_gem == "carnelian" else "거짓")

#이러한 조건문을 삼항연산자를 활용한 조건문이라 부른다.
```

```python
num = int(input('정수 입력 : '))
print('홀수' if num % 2 != 0 else '짝수')
```

```python
a = int(input('첫번째 정수 입력 : '))
b = int(input('두번째 정수 입력 : '))

big = a if a > b else b
small = a if a < b else b
print(big, ">", small)
```

```python
a = int(input('첫번째 정수 입력 : '))
b = int(input('두번째 정수 입력 : '))

if a >= b:
    print(a, "≥", b)

else:
    print(a, "<", b)
```

match 구문을 활용해도 조건문을 작성할 수 있다. 이 때 match는 범위를 지정하는 것은 어렵지만 

케이스에 정확히 일치하는지 검증하는 것에는 용이하다.

```python
print("=== OhGiraffers vending machine ===")
print("tiger: 우리는 1인 24시간 교대제로 간다")
print("monkey: 24시간 문의 환영, 다른 말로 하면?")
print("squirrel: 저도 하루에 너다섯 시간 자는데요?")
print("rabbit: 살아남는 사람이 강한 사람입니다.")
print("====================================")
print("만나고 싶은 선생님을 입력해 주세요")
input_teacher = input()
print("====================================")

sleep_time = 0
match input_teacher:
    case "tiger":
        print("호랑이 선생님의 메타버스에 오신 것을 환영합니다.")
        sleep_time = 3
    case "monkey":
        print("원숭이 선생님의 밀림에 오신 것을 환영합니다.")
        sleep_time = 3.5
    case "squirrel":
        print("다람쥐 선생님의 매력의 늪에 오신 것을 환영합니다.")
        sleep_time = 5
    case "rabbit" :  
        print("토끼 선생님의 다정보스 랩에 오신 것을 환영합니다.")
        sleep_time = 5.2
    case _:
        print("공부할 마음이 없으신 건 아니죠?")

print("매일 " + str(sleep_time) + "시간의 수면이 예상됩니다.")
```
