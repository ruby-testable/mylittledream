분기문은 조건문과 함께 작성되어 반복문의 무한 실행을 방지하는 역할을 담당한다.

```python
for x in [1, 2, 3]:
    
    print("앞", x)
    
    if x == 2:
        print("if", x)
        print("끝")  	
        break
    
    print("뒤", x)

#break로 인해 x가 2일때의 조건문을 실행함과 동시에 반복을 멈추었다.
```

```python
for x in range(3):	
    print("앞", x)
    
    if x == 1:
        print("if", x)
        print("continue 등장!")  	
        continue
    
    print("뒤", x)

"""
continue를 만나면 해당 반복문의 시퀀스를 종료시킨다.
예를 들어, 이 예시에서는 x = 1 일때의 시퀀스가 종료되고 
x = 2일때의 시퀀스로 넘어갔다.
"""
```

```python
for i in range(3):

    for j in range(3):
        print(i,j)
        break

#break는 해당 카테고리의 반복만 멈출뿐, 상위 카테고리의 작동에 간섭하지 않는다.
```
