모듈은 파이썬 파일 하나를 의미한다. 이 파일에서 필요한 정보(함수, 클래스, 변수 등)를 가져와서 사용할 수 있다. 모듈의 의의는 재사용에 있다고 볼 수 있을 것이다.

```python
import ruby_module

print(ruby_module.gems)
```

모듈 내의 함수나 메서드, 클래스를 별개로 도입할 수도 있다. 이 경우는 명령어의 작성이 비교적 간편해 진다.

```python
from ruby_module import emerald, gems

print(gems)
print(emerald())
```

모듈의 이름을 별칭으로 바꿀 수도 있다. 일반적으로 코드의 길이를 줄이기 위해 활용한다.

```python
import ruby_module as RBModule

print(RBModule.gems)
```

모듈 내의 함수, 메서드, 클래스도 별칭을 부여해 사용할 수 있다.

```python
from ruby_module import sapphire as blue, emerald as green

print(blue())
print(green())
```

표준 라이브러리 모듈은 별개의 파일 생성 없이 import 가능하다.

```python
# math 모듈 예시
import math

print(math.sqrt(16))        # 출력: 4.0
print(math.factorial(5))    # 출력: 120
print(math.pi)              # 출력: 3.141592653589793
```
