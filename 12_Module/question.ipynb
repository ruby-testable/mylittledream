문제 1️⃣: 짝수만 뽑아서 합 구하기

리스트가 주어졌을 때, 짝수만 골라서 합을 구하는 프로그램을 작성하시오.

```python
numbers = [1, 2, 3, 4, 5, 6]

def sum_even(list):
    even_list = [even for even in list if even % 2 == 0]
    return sum(even_list)

print(sum_even(numbers))
```

문제 1️⃣: 리스트에서 중복 제거 후 정렬

숫자가 담긴 리스트가 주어졌을 때,
중복된 숫자를 제거하고 오름차순으로 정렬한 리스트를 반환하는 함수를 작성하세요.

```python
numbers = [4, 2, 5, 2, 3, 4, 1]

def well_ordering(lst): 
    set1 = set(lst)
    new_lst = list(set1)
    new_lst.sort()
    return new_lst

print(well_ordering(numbers))
```

```python
numbers = [4, 2, 5, 2, 3, 4, 1]

def well_ordering(lst):
    return sorted(set(lst))

print(well_ordering(numbers))
```

문제 2️⃣: 팰린드롬 판별

문자열이 주어졌을 때,
**앞뒤로 읽어도 같은 문자열(팰린드롬)**인지 확인하는 함수를 작성하세요.

```python
s = "level"
t = "sapphire"

def palindrome(word):
    reversed_word = str(word[::-1])
    return word == reversed_word
      

print(palindrome(s))
print(palindrome(t))
```

4. 소수 판별

- 숫자를 입력받아 소수인지 판별하는 함수 작성

- 1~100 사이 소수를 모두 리스트로 반환

```python
# 소수 판별 함수
def is_prime(n):
    if n < 2:  # 2보다 작은 수는 소수가 아님
        return False
    for i in range(2, int(n**0.5) + 1):  # 2부터 sqrt(n)까지 확인
        if n % i == 0:
            return False
    return True

# 1~100 사이 모든 소수 출력
primes = [x for x in range(1, 101) if is_prime(x)]
print(primes)
```
