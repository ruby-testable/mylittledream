ğŸ”¹ ë¬¸ì œ 3) ì˜í™” í‰ì  ì‹œìŠ¤í…œ + ìƒì†

Movie í´ë˜ìŠ¤ì™€ AnimatedMovie í´ë˜ìŠ¤ í™œìš©:

1. Movie: ì œëª©, ê°ë…, í‰ì  â†’ info() "ì œëª© - ê°ë… (í‰ì : x.x)"

```python
class Movie:

    def __init__(self, title, director, rating):
        self.title = title
        self.director = director
        self.rating = rating

    def info(self):
        return str(f"{self.title} - {self.director} (í‰ì : {self.rating})")

class AnimatedMovie(Movie):

    def __init__(self, title, director, rating, limit):
        super().__init__(title, director, rating)
        self.limit = limit

    def info(self):
        return str(f"{self.title} - {self.director} (í‰ì : {self.rating}, ì—°ë ¹ ì œí•œ: {self.limit}ì„¸)")

    
movies = [
    Movie("ì¸ì…‰ì…˜", "ë†€ë€", 8.8),
    AnimatedMovie("í† ì´ ìŠ¤í† ë¦¬", "ì¡´ ë¼ì„¸í„°", 9.1, 0),
    Movie("ìŠˆë ‰", "ì• ë¤ìŠ¨", 7.8),
    AnimatedMovie("ê²¨ìš¸ì™•êµ­", "í¬ë¦¬ìŠ¤ ë²…", 8.5, 6)
]

nice_movies = []

for movie in movies:
    
    if isinstance(movie, AnimatedMovie):
        if movie.limit == 0:
            nice_movies.append(movie.title)

    else:
        nice_movies.append(movie.title)

print(nice_movies)


```

```python
a = [1,2,3,4]
b = [3,4,5,6]
# ê³µí†µ ìš”ì†Œ ì¶œë ¥ â†’ [3,4]
# aë§Œ ìˆëŠ” ìš”ì†Œ ì¶œë ¥ â†’ [1,2]

def intersection(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    return set1 & set2

def difference(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    return set1 - set2

print(intersection(a, b))
print(difference(a, b))
```
