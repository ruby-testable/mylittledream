ë‹¤ìŒ ì¡°ê±´ì„ ë§Œì¡±í•˜ëŠ” Streamlit í˜ì´ì§€ `pages/1_input_page.py`ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

    - ë‹‰ë„¤ì„ì„ ì…ë ¥ë°›ëŠ” `st.text_input` (ë ˆì´ë¸”: `ë‹‰ë„¤ì„ ì…ë ¥`)

    - êµ­ì ì„ ì„ íƒí•˜ëŠ” `st.selectbox`
        - ì˜µì…˜: `['í•œêµ­', 'ì¤‘êµ­', 'ì¼ë³¸', 'ë¯¸êµ­', 'ì™¸ê³„']`
    - ì·¨ë¯¸ë¥¼ ì„ íƒí•˜ëŠ” `st.radio`
        - ì˜µì…˜: `['ë§›ì§‘ íƒë°©', 'ì˜í™” ê°ìƒ', 'ìŒì•… ê°ìƒ', 'ëœ¨ê°œì§ˆ']`
    - ê°œì¸ì •ë³´ ì œê³µ ë™ì˜ ì—¬ë¶€ë¥¼ ì²´í¬í•˜ëŠ” `st.checkbox` (ë ˆì´ë¸”: `ê°œì¸ì •ë³´ ì œê³µ ë™ì˜`)
3. ì˜¤ë¥¸ìª½ ì»¬ëŸ¼(col2)ì—ëŠ” `"âœ… ì…ë ¥ ì™„ë£Œ"` ë¼ëŠ” ë²„íŠ¼ì„ í•˜ë‚˜ ë§Œë“¤ê³ , ë²„íŠ¼ì´ ëˆŒë¦¬ë©´ ë‹¤ìŒ ë‚´ìš©ì„ `st.write()`ë¡œ ì¶œë ¥í•˜ì„¸ìš”.
    - `ì´ë¦„ì´ ë­ì•¼? {ë‹‰ë„¤ì„}`
    - `ëª‡ ì‚´ì´ì•¼? {ë‚˜ì´}`
    - `ì–´ë””ì„œ ì™”ì–´? {êµ­ì }`
    - `ì·¨ë¯¸ê°€ ë­ì•¼? {ì·¨ë¯¸}`

```python
import streamlit as st

st.title("ì‚¬ìš©ì ì…ë ¥ ë°›ëŠ” í˜ì´ì§€")

col1, _, col2 = st.columns(3)

with col1:
    nickname = st.text_input("ë‹‰ë„¤ì„ ì…ë ¥")
    age = st.text_input("ë‚˜ì´ ì…ë ¥")
    country = st.selectbox("êµ­ì ", ['í•œêµ­', 'ì¤‘êµ­', 'ì¼ë³¸', 'ë¯¸êµ­', 'ì™¸ê³„'])
    hobby = st.radio("ì·¨ë¯¸", ['ë§›ì§‘ íƒë°©', 'ì˜í™” ê°ìƒ', 'ìŒì•… ê°ìƒ', 'ëœ¨ê°œì§ˆ'])
    privacy = st.checkbox("ê°œì¸ì •ë³´ ì œê³µ ë™ì˜")


with col2:
    if st.button("âœ… ì…ë ¥ ì™„ë£Œ~"):
        st.write(f"ì´ë¦„ì´ ë­ì•¼? {nickname}")
        st.write(f"ëª‡ ì‚´ì´ì•¼? {age}")
        st.write(f"ì–´ë””ì„œ ì™”ì–´? {country}")
        st.write(f"ì·¨ë¯¸ê°€ ë­ì•¼? {hobby}")
        st.write(f"ê°œì¸ì •ë³´ ì œê³µì— ë™ì˜í•´? {privacy}")
```
