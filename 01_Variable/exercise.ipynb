언더스코어는 가장 최근에 실행한 값을 출력한다. 예를 들어, 10을 마지막으로 출력했으면 10을 반영하고, ruby를 마지막으로 출력했으면 ruby를 반영한다.

```python
print(_)
print(_ * 3)
```

언더스코어는 특정 값을 무시하고 싶을 때도 사용된다. + *(애스터리스크)는 코딩 문법 상 모든 것을 의미한다.

```python
x, _, z = (1, 2, 3)
print(x)
print(z)
```

```python
x, *_, z = (1, 2, 3, 4, 5)
print(x)
print(z)
```

큰 숫자의 자릿수 구분을 위해 사용할 수도 있다. 이때의 언더스코어는 읽지 않는 것처럼 보인다.
