논리자료형은 변수 작성시 첫 글자를 대문자로 작성해야한다.

```python
bool1 = True
bool2 = False

print(type(bool1))
print(type(bool2))
```

```python
# ==은 좌항과 우항의 값이 동일한지 비교한다. = 한개만 사용된 것은 변수 저장의 목적으로
# 사용되고 있으므로 이렇게 하는 듯. !는 컴퓨터 언어에서 부정의 의미를 나타낸다.

print(1 == 3)                            # False
print('ruby' != 'Ruby')                  # True
print(100 > 99)                          # True
print(100 < 99)                          # False
print(0 >= 1)                            # False
print(0 <= 1)                            # True
```

```python
# 대문자는 아스키코드표 상에서 소문자에 비해 작은 수를 할당 받았다.

print("a" > "A")
```

and 연산자는 논리 자료형 전부가 True여야 True 결과를 출력한다.

or 연산자는 논리 자료형 중 하나 이상 True여야 True 결과를 출력한다.

```python
print(1 != 3 or 100 > 99 or True)        
print(1 != 3 or 100 > 99 or 100 < 99)    
print(1 == 3 or False or 100 < 99) 
```

strip 메서드는 문자 양 끝의 제시된 문자를 제거한다.

lstrip은 왼쪽만, rstrip은 오른쪽만 제거한다.

```python
origin = "ruby"
with_white_space = "      ruby      "

print(with_white_space.strip())
```

```python
origin = "ruby"
with_white_space = "      ruby      "

print(with_white_space.strip("r "))

"""
두 글자 이상의 문자를 strip() 괄호 안에 입력 시 그 중 하나가 조건에 맞으면 삭제,
조건에 맞지 않을 경우 실행을 멈추는 것으로 보임.
"""
```

```python
origin = "ruby"
with_white_space = "      ruby      "

print(origin.strip("ub"))

#ub가 ruby의 양 끝이 아니므로 문자열의 제거가 일어나지 않는다.
```

```python
origin_str = 'hELLO wORLD!'

print(origin_str.upper())         # HELLO WORLD!
print(origin_str.lower())         # hello world!
print(origin_str.capitalize())    # Hello world!
print(origin_str.swapcase())      # Hello World!
print(origin_str.title())         # Hello World!
```
